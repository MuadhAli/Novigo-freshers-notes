## ⚙️ Section 3: Functions — Standalone and Class-Based

Both patterns are essential. Standalone for simple utilities. Classes for grouping related logic — agents, tools, memory managers.


````python
a, b = 10, 2
````

````python
# ══════════════════════════════════════════════════════════════════
# FUNCTION WITHOUT A CLASS
# Standalone — call it directly, no object needed
# ══════════════════════════════════════════════════════════════════

def add_numbers(a: float, b: float) -> float:
    # (a: float, b: float) -> float means:
    # accepts two floats, returns a float
    return a + b

# print (add_numbers(10, 2))
````

````python
# ══════════════════════════════════════════════════════════════════
# FUNCTION INSIDE A CLASS
# Grouped, reusable — in AI: classes group Agents, Tools, Memory etc.
# ══════════════════════════════════════════════════════════════════

class MathOperations:
    # Groups all math-related functions together
    # 'self' = reference to this particular instance of the class

    def add_numbers(self, a: float, b: float) -> float:
        # Same logic as standalone, but now it belongs to a class
        return a + b

    def describe(self) -> str:
        return "MathOperations — groups: add, subtract, multiply, divide"

# Step 1: instantiate — create an object FROM the class blueprint
math_obj = MathOperations()

# Step 2: call the method on the object
result = math_obj.add_numbers(8, 4)
print(f"[Class] math_obj.add_numbers(8, 4) = {result}")
print(f"Description: {math_obj.describe()}")
````

### THE LOGIC CELL
###   a, b  → declared as variables
###   total → sum using add_numbers()
###   if total is EVEN  → multiply: total * b
###   if total is ODD   → divide:   total / b
###   print answer in a right-angled triangle

````python
# Step 1: Variables
a = 8
b = 4

# Step 2: Get sum using our function
total = add_numbers(a, b)

print(f"a           = {a}")
print(f"b           = {b}")
print(f"total (a+b) = {total}")
print(f"Is even?    = {total % 2 == 0}   (total % 2 = {total % 2})")
````

````python

# Step 3: Even → Multiply | Odd → Divide
if total % 2 == 0:
    answer = total * b
    label  = f"{total} x {b} = {answer}  [EVEN → Multiply]"
else:
    answer = round(total / b, 2)
    label  = f"{total} / {b} = {answer}  [ODD  → Divide]"

print(f"\nOperation : {label}")
print(f"Answer    : {answer}")

````

````python
# Step 4: Print answer in a right-angled triangle
print(f"\n📐 Right-Angled Triangle  (answer = {answer}):")
print()

ROWS = 5
for i in range(1, ROWS + 1):
    row_str = "★ " * i
    if i == ROWS:
        print(f"{row_str}  ◄── answer: {answer}")   # last row shows the result
    else:
        print(row_str)
````

````python
import requests

def call_math_api(operation: str, a: float, b: float = None) -> dict:
    # Calls the UV FastAPI math server — synchronous (blocking)
    # operation : 'add', 'subtract', 'multiply', 'divide', 'sqrt'
    # b is optional — sqrt only needs a
    url     = f"{BASE_URL}/{operation}"
    headers = {"X-API-Key": API_KEY, "Content-Type": "application/json"}
    payload = {"a": a}
    if b is not None:
        payload["b"] = b

    try:
        resp = requests.post(url, json=payload, headers=headers, timeout=5)
        resp.raise_for_status()    # raises HTTPError for any 4xx or 5xx status
        return resp.json()

    except requests.exceptions.ConnectionError:
        return {"error": "Cannot connect — is your UV server running?"}
    except requests.exceptions.Timeout:
        return {"error": "Timed out after 5 seconds"}
    except requests.exceptions.HTTPError as e:
        return {"error": f"HTTP {e.response.status_code}: {e.response.text}"}


````

## 🧬 Section 8: JSON + Pydantic Validation

### Why this is critical for AI development

```
LLM Output (raw text/JSON)
        ↓ json.loads()
    Python dict
        ↓ Pydantic Model(**dict)
  Validated Object  ← crashes here if bad, not 3 steps later
        ↓
  Your agent code (safe to use)
```

**Without Pydantic:** a missing field or wrong type from the LLM = crash somewhere unexpected.  
**With Pydantic:** bad data is caught immediately with a clear, actionable error.


````python
import json

# ── JSON string → Python dict (json.loads) ────────────────────────────────
raw_json = '{"operation": "add", "a": 10, "b": 5, "result": 15, "success": true}'

print(f"Raw type   : {type(raw_json)}")
print(f"Raw value  : {raw_json}")

data = json.loads(raw_json)     # parse string → dict
print(f"\nParsed type: {type(data)}")
print(f"Parsed val : {data}")
# print(f"result key : {data['result']}  (type: {type(data['result']).__name__})")
````

````python
# ── Python dict → JSON string (json.dumps) ────────────────────────────────
agent_output = {
    "agent":        "MathAgent",
    "steps":        ["parsed input", "called API", "validated response"],
    "final_answer": 15,
    "confidence":   0.99
}

json_string = json.dumps(agent_output, indent=2)  # indent=2 for pretty print
print("Dict → JSON string:")
print(json_string)

# indent=None → compact (use for API payloads)

````

````python
compact = json.dumps(agent_output)
print(f"\nCompact: {compact}")
````

````python
from pydantic import BaseModel
from typing import Optional
````

````python
class MathRequest(BaseModel):
    # What we SEND to the server — validated before sending
    a: float
    b: Optional[float] = None   # Optional: sqrt only needs 'a'

class MathResponse(BaseModel):
    # What we EXPECT back — validated on receipt
    operation: str
    result:    float
    success:   bool = True      # defaults to True if not in response

# ── Valid data ────────────────────────────────────────────────────────────
good = {"operation": "add", "result": 15.0, "success": True}
resp = MathResponse(**good)     # ** unpacks dict as keyword arguments
````

````python
print(resp)
````

````python
print("=== Valid Response ===")
print(f"Parsed     : {resp}")
print(f"result     : {resp.result}")
print(f"As dict    : {resp.model_dump()}")
print(f"As JSON    : {resp.model_dump_json()}")
````

````python
# ── Auto type coercion ────────────────────────────────────────────────────
print("=== Type Coercion — Pydantic auto-converts compatible types ===")
coerced = {"operation": "multiply", "result": "42", "success": "true"}
resp2   = MathResponse(**coerced)

print(resp2)
# print(f"  '42' (str)   →  {resp2.result} (float)  ✅")
# print(f"  'true' (str) →  {resp2.success} (bool)    ✅")
````

````python

# ── Invalid data ──────────────────────────────────────────────────────────
from pydantic import ValidationError

print("\n=== Invalid Data — Pydantic stops this from crashing your agent ===")

bad_cases = [
    ("Wrong type — can't convert",   {"operation": "add",  "result": "not-a-number"}),
    ("Missing required field",       {"result": 15.0}),
    ("Empty dict",                   {}),
]

for label, bad_data in bad_cases:
    try:
        MathResponse(**bad_data)
    except ValidationError as e:
        print(f"\nCase : {label}")
        print(f"Input: {bad_data}")
        for err in e.errors():
            field = err['loc'][0] if err['loc'] else 'unknown'
            print(f"  ❌ field='{field}' | {err['msg']}")
````

## 🛡️ Section 9: Error Handling — `try / except / finally`

In AI systems, failures are **expected**. Design for them:

| Failure Mode | Exception |
|---|---|
| Server not running | `ConnectionError` |
| Slow server | `Timeout` |
| Bad status code (401, 500) | `HTTPError` |
| Server returned non-JSON | `JSONDecodeError` |
| Missing/wrong field from LLM | `ValidationError` (Pydantic) |
| Divide by zero | `HTTPError` from server |

**`finally`** always runs — use it for logging, cleanup, and tracing.


````python
import json
import requests
from pydantic import ValidationError
from typing import Optional
````

````python
def safe_math_call(operation: str, a: float, b: float = None) -> Optional[MathResponse]:
    # Fully protected API call: HTTP + JSON parse + Pydantic validation
    # Returns MathResponse on success, None on any failure — never crashes
    print(f"\n→ Calling '{operation}'  a={a}  b={b}")
    try:
        # ── 1. Build and send the HTTP request ────────────────────
        url     = f"{BASE_URL}/{operation}"
        headers = {"X-API-Key": API_KEY}
        payload = {"a": a}
        if b is not None:
            payload["b"] = b

        resp = requests.post(url, json=payload, headers=headers, timeout=5)
        resp.raise_for_status()     # raises HTTPError for 4xx / 5xx

        # ── 2. Parse the JSON body ─────────────────────────────────
        raw = resp.json()
        print(f"  Raw response : {raw}")

        # ── 3. Validate with Pydantic ──────────────────────────────
        validated = MathResponse(**raw)
        print(f"  ✅ Validated!  result = {validated.result}")
        return validated

    except requests.exceptions.ConnectionError:
        print("  ❌ ConnectionError — is the UV server running?")

    except requests.exceptions.Timeout:
        print("  ❌ Timeout — server took > 5 seconds")

    except requests.exceptions.HTTPError as e:
        print(f"  ❌ HTTPError {e.response.status_code}: {e.response.text}")

    except json.JSONDecodeError as e:
        print(f"  ❌ JSONDecodeError — bad response body: {e}")

    except ValidationError as e:
        print("  ❌ Pydantic ValidationError:")
        for err in e.errors():
            field = err['loc'][0] if err['loc'] else 'unknown'
            print(f"     field='{field}': {err['msg']}")

    except Exception as e:
        print(f"  ❌ Unexpected: {type(e).__name__}: {e}")

    finally:
        # Always runs — in production: log to your tracing/monitoring system
        print(f"  [log] safe_math_call('{operation}') — finished")

    return None   # only reached if an exception was caught above
````

````python
# ── Run Tests ─────────────────────────────────────────────────────────────
print("=== Test 1: Normal call ===")
r = safe_math_call("add", 10, 5)

print("\n=== Test 2: Divide by zero ===")
r = safe_math_call("divide", 10, 0)

print("\n=== Test 3: Server unreachable ===")
try:
    requests.post("http://127.0.0.1:9999/add", json={"a": 1, "b": 2}, timeout=1)
except requests.exceptions.ConnectionError:
    print("  ❌ ConnectionError (expected — wrong port, no server there)")
````

## ⚡ Section 10: Async Python — Coroutines & Concurrent Calls

### Why async matters for AI agents

```
SYNC  (blocking):    [add]──►[subtract]──►[multiply]──►[divide]──►[sqrt]   = 5s total
                     
ASYNC (concurrent):  [add]      ─┐
                     [subtract]  │
                     [multiply]  ├──► all done in ~1s (= slowest single call)
                     [divide]    │
                     [sqrt]     ─┘
```

**Key terms:**
- `async def` — defines a **coroutine** (a pausable, non-blocking function)
- `await` — pause here and give control back; resume when result is ready
- `asyncio.gather()` — fire multiple coroutines **at the same time**
- `httpx.AsyncClient` — async HTTP client (like `requests`, but non-blocking)


````python
import asyncio

# ── A simple coroutine ────────────────────────────────────────────────────
async def greet(name: str, delay: float) -> str:
    # async def makes this a coroutine — it must be awaited to run
    print(f"  [{name}] starting...")
    await asyncio.sleep(delay)    # non-blocking: other tasks run during this wait
    msg = f"{name} finished after {delay}s"
    print(f"  [{name}] done")
    return msg

# Jupyter already has an event loop — use top-level await directly
print("=== Single coroutine ===")
result = await greet("Junior Dev", 0.3)
print(f"Returned: {result}")
````

````python
import time

# ── Sequential await (sync-style) vs asyncio.gather (concurrent) ──────────
print("=== Sequential (like sync) ===")
start = time.time()
await greet("Alice", 0.3)
await greet("Bob",   0.5)
await greet("Carol", 0.4)
elapsed = time.time() - start
print(f"Total: {elapsed:.2f}s  (expected ~1.2s = 0.3+0.5+0.4)\n")

print("=== Concurrent with asyncio.gather ===")
start = time.time()
results = await asyncio.gather(
    greet("Alice", 0.3),
    greet("Bob",   0.5),
    greet("Carol", 0.4),
)
elapsed = time.time() - start
print(f"\nTotal: {elapsed:.2f}s  (expected ~0.5s = max of all delays)")
print(f"Results: {results}")
````

````python
import httpx
import asyncio
import time

# NOTE: BASE_URL and API_KEY are set in Section 7 — run that cell first

async def async_math_call(
    client: httpx.AsyncClient,
    operation: str,
    a: float,
    b: float = None
) -> dict:
    # Async version of our math API caller
    # Uses httpx.AsyncClient — the async counterpart of requests.Session
    url     = f"{BASE_URL}/{operation}"
    headers = {"X-API-Key": API_KEY}
    payload = {"a": a}
    if b is not None:
        payload["b"] = b

    try:
        resp = await client.post(url, json=payload, headers=headers, timeout=5.0)
        resp.raise_for_status()
        data = resp.json()
        print(f"  ✅ {operation:<12} → {data}")
        return data
    except Exception as e:
        err = {"operation": operation, "error": str(e)}
        print(f"  ❌ {operation:<12} → {err}")
        return err


async def run_all_operations():
    # asyncio.gather fires ALL coroutines at the same time
    # httpx.AsyncClient is shared — enables connection pooling
    print("🚀 Firing all 5 operations concurrently...\n")

    async with httpx.AsyncClient() as client:
        results = await asyncio.gather(
            async_math_call(client, "add",      a=20, b=5),
            async_math_call(client, "subtract", a=20, b=5),
            async_math_call(client, "multiply", a=20, b=5),
            async_math_call(client, "divide",   a=20, b=5),
            async_math_call(client, "sqrt",     a=225),
        )

    print(f"\n📊 All {len(results)} results:")
    for r in results:
        op  = r.get("operation", "?")
        res = r.get("result", r.get("error", "?"))
        print(f"   {op:<12} → {res}")
    return results

# ── Run it ────────────────────────────────────────────────────────────────
print("=== Concurrent Math API Calls ===")
start       = time.time()
all_results = await run_all_operations()
elapsed     = time.time() - start
print(f"\n⏱  Total: {elapsed:.3f}s  (all concurrent — not sequential!)")
````
